```python
import random

def tanh(x):
    return (2 / (1 + np.exp(-x))) - 1

def tanh_dy_dx(x):
    sech_x =  (2/(np.exp(x)+np.exp(-x))) 
    return sech_x*sech_x

class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
 
    def feedforward(self, inputs):
        # Вводные данные о весе, добавление смещения 
        # и последующее использование функции активации
 
        total = np.dot(self.weights, inputs) + self.bias
        return tanh(total)
        
    def correct_weights(self, inputs, err, std_train):
        self.weights[0]=self.weights[0] + err*inputs[0]*std_train
        self.weights[1]=self.weights[1] + err*inputs[1]*std_train
        
        
    
class NeuralNetwork:    
    def __init__(self):
        beas = 0;
        weights = np.array([random.uniform(-1,1), random.uniform(-1,1)])
        self.h1 = Neuron(weights, beas)
        weights = np.array([random.uniform(-1,1), random.uniform(-1,1)])
        self.h2 = Neuron(weights, beas)
        weights = np.array([random.uniform(-1,1), random.uniform(-1,1)])
        self.o1 = Neuron(weights, beas)
    
    
    def feedforward(self, x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
 
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2]))
 
        return out_o1
    
    def feedforward_lerning(self, x, expected):
        
        
        
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        o1_input = np.array([out_h1, out_h2]);
        out_o1 = self.o1.feedforward(o1_input)
    
        err_o1 = tanh_dy_dx(out_o1)
        err_o1 = err_o1 * (expected-out_o1)
        
        err_h1_h2 = self.o1.weights*err_o1*(np.array([tanh_dy_dx(out_h1),tanh_dy_dx(out_h2)]))
        
        std_train = 0.1  
        
        self.h1.correct_weights(x, err_h1_h2[0], std_train)        
        self.h2.correct_weights(x, err_h1_h2[1], std_train)
        self.o1.correct_weights(o1_input, err_o1, std_train)
        
        
net = NeuralNetwork()
```

```python

net = NeuralNetwork()

error=0
cicles = 4
i = 0
while i < cicles:
    x1=i%2;
    x2 = (1+i//2)%2
    print("x1="+str(x1) + ", x2=" + str(x2))
    expected = 1 if x1==1 and x2 == 0 else 0;    
    y = net.feedforward(np.array([x1,x2]));
    error+= (y-expected)*(y-expected)
    i+=1;
error/=cicles;
print(error)

```
