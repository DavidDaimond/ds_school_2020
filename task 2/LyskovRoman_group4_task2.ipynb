# DECISION TREE

```python
# selecting the target variable
y = df['Churn'].astype('int')
X = df.drop('Churn', axis=1)
```

```python
X.shape, y.shape
```

```python
# split our data: training and test samples using scikit-learn
X_train, X_valid, y_train, y_valid = train_test_split(X,y,
                                                     test_size = 0.3,
                                                     random_state = 26)
```

```python
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

```python
tree_model = DecisionTreeClassifier(random_state = 26)
```

```python
# average quality on crossvalidation
np.mean(cross_val_score(tree_model, X_train, y_train, cv=5))
```

```python
tree_params = {'max_depth': np.arange(1, 11), 'max_features':[.5, .7, 1]}
```

```python
tree_grid = GridSearchCV(tree_model, tree_params, cv=5, n_jobs=-1)
```

```python
%%time
tree_grid.fit(X_train, y_train)
```

```python
# best maximum depth = 4
# best maximum number of signs = 0.7
tree_grid.best_score_, tree_grid.best_params_
```

```python
decision_tree = DecisionTreeClassifier(max_depth = 4, max_features = 0.7)
```

```python
decision_tree.fit(X_train, y_train)
```

```python
add_accuracy("Decision Tree", decision_tree)
```
